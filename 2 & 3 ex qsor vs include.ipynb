```python
def Qsort(arr, L, R, s, c):
    if len(arr)>-1:
        i = L
        j = R
        x = arr[round((L+R)/2)]
        while (i<=j):
            while (arr[i]<x and i<=j):
                i+=1
                c+=1
            if i>j:
                c+=1
            while (arr[j]>x and i<=j):
                j-=1
                c+=1
            if i>j:
                c+=1
            if i<=j:
                if i!=j:
                    arr[i], arr[j] = arr[j], arr[i]
                    s+=1
                i+=1
                j-=1
        if (L<j):
            s,c = Qsort(arr, L, j, s, c)
        if (i<R):
            s,c = Qsort(arr, i, R, s, c)    
    return s, c
```

```python
def insertion(array):
    c = 0
    s = 0
    for i in range(1,len(array)):
        j=i
        while j>0 and array[j]<array[j-1]:
            array[j], array[j-1] = array[j-1], array[j]
            s+=1
            c+=1
            j-=1
        if array[j]>=array[j-1] or j<=0:
            c+=1
    return s, c
```

# Задача 3(сортировка подсчетом) 

```python
def Counting_Sort(A):
    B = [0 for i in range(len(A))]
    k = max(A)
    n = len(A)
    C = [0 for i in range(k+1)]
    for i in range(n):
        C[A[i]] +=1
    for i in range(1,k):
        C[i] = C[i] + C[i-1]
    for i in range(n-1,-1,-1):
        C[A[i]] -= 1
        B[C[A[i]]] = A[i]
```

```python
a = [1,2,6,1,1,2,3]
Counting_Sort(a)
```
